```python
import os
import time
from glob import glob

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import audb
import audiofile
import opensmile
```

```python



db = audb.load(
    'emodb',
    version='1.1.1',
    format='wav',
    mixdown=True,
    sampling_rate=16000,
    media='wav/03a01.*',  # load subset
    full_path=False,
    verbose=False,
)

files = glob('Dataset\Audio_Speech_Actors_01-24\*\*.wav')
print(files)

signal, sampling_rate = audiofile.read(
    files[0],
    duration=10,
    always_2d=True,
)

smile = opensmile.Smile(
    feature_set=opensmile.FeatureSet.eGeMAPSv02,
    feature_level=opensmile.FeatureLevel.Functionals,
)


smile.process_signal(
    signal,
    sampling_rate
)
```
